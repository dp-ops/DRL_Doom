2. **Activate the environment**

4. **Install required packages**

The following cell contains commands for these steps. Run them in your terminal/command prompt, not in this notebook.

```python
# These commands should be run in your terminal/command prompt, not in this notebook
# They are provided here for reference

# 1. Create a virtual environment
# Windows
# python -m venv doom_env

# macOS/Linux
# python3 -m venv doom_env

# 2. Activate the environment
# Windows
# doom_env\Scripts\activate

# macOS/Linux
# source doom_env/bin/activate

# 3. Verify and update pip
# python -m pip --version
# python -m pip install --upgrade pip

# 4. Install required packages
# python -m pip install numpy torch gym vizdoom

# Create requirements.txt file (optional)
# pip freeze > requirements.txt

# To deactivate the virtual environment when done
# deactivate
```
